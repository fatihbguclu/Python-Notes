<div class="alert alert-info">
<b>ÇİZGİ GRAFİKLERİ (LINE GRAPH)</b> 
</div>

<div class="alert alert-success">
Bu tip grafikler en basit grafik tipi olup iki değişken arasındaki bağlantıyı görselleştirmemize yaramaktadır. Çizgi grafikler adından anlaşılabileceği gibi birbirini takip eden veriler arasına çizgi oluşturulmasını sağlar. Bunu gösterebilmek için basit bir fonksiyon kullanalım
</div>

En basit haliyle bir "sinüs" verisi yaratalım ve bunun çizgi grafiğini gösterelim. İlerleyen adımlarda bu grafiğin üzerinde değişiklikler yaparak grafiği geliştirelim.

````python
# 0'dan 2*pi'ye
x = np.linspace(0,np.pi*2,50)
````

````python
# y eksenindeki veriyi de oluşturalım
y = np.sin(x)
````

````python
plt.figure()
plt.plot(x,y)
plt.show();
````

Gördüğünüz gibi bu grafik bile çok şey anlatmakta fakat bunu biraz daha geliştirelim hep beraber. Adım adım ilerliyelim: <br>
* Grafiğe eksen isimlerini ve başlığı ekleyelim. 
* Aynı zamanda ızgara da (grid) ekleyelim. 
<br> 

````python
plt.figure(figsize=[13,7])
plt.plot(x,y)

# y eksenine isim ekliyoruz
plt.ylabel("y değişkeni");
# x eksenine isim ekliyoruz
plt.xlabel("x değişkeni")
# grafiğe başlık ekliyoruz
plt.title("Örnek çizgi grafiği (sin)");

# ızgara ekliyoruz
plt.grid()

plt.show();
````

* Grafiğimizi daha anlaşılır hale getirmek için veri noktalarını da belirginleştirelim. 
* Rengi kırmızı olan ikinci bir çizgi grafiği ekleyelim. Çizginin kalınlığı 5 olsun
* Ayrıca bir de lejand (legend) ekleyim

````python
plt.figure(figsize=fig_size)

# Veri noktalarını belirginleştiriyoruz.
plt.plot(x,y,'-ob')
# İkinci bir grafik ekleyelim. Kesik çizgili ve kırmızı olsun
plt.plot(x,y*0.3,'--r',linewidth=5)


# y eksenine isim ekliyoruz
plt.ylabel("y değişkeni")
# x eksenine isim ekliyoruz
plt.xlabel("x değişkeni")
# grafiğe başlık ekliyoruz
plt.title("Örnek çizgi grafiği")
# ızgara ekliyoruz
plt.grid()
# lejand ekliyoruz
plt.legend(["Normal sinus","0.3 ile çarpılmış"])

plt.show();
````

* x eksenin çizgilerini 0, pi/2, pi, 3pi/2, 2pi olacak şekilde ayarlayalım ve bunların isimlerini x ekseni üzerinde gösterelim.
* Ayrıca grafiğin üzerinde pi değerine bir annotasyon koyalım.

````python
plt.figure(figsize=fig_size)

# Veri noktalarını belirginleştiriyoruz.
plt.plot(x,y,'-ob')
# İkinci bir grafik ekleyelim. Kesik çizgili ve kırmızı olsun
plt.plot(x,y*0.3,'--r',linewidth=5)


# y eksenine isim ekliyoruz
plt.ylabel("y değişkeni")
# x eksenine isim ekliyoruz
plt.xlabel("x değişkeni")
# grafiğe başlık ekliyoruz
plt.title("Örnek çizgi grafiği")
# ızgara ekliyoruz
plt.grid()
# lejand ekliyoruz
plt.legend(["Normal sinus","0.3 ile çarpılmış"]);

# Lokasyonları istediğimiz değerlere ayarlayalım. Bu değerlerin yazı kısmını ise biraz Latex kullanarak yazabiliriz. 
plt.xticks([0,np.pi/2,np.pi,1.5*np.pi,2*np.pi],["0",r"$\dfrac{\pi}{2}$","$\pi$",r"$\dfrac{3\pi}{2}$","$2\pi$"]);

plt.annotate("Tam bu noktaya bakın!\n ($\pi$,0)",[np.pi,0],[1.5*np.pi,0.6],
             arrowprops=dict(facecolor='m', shrink=0.05),fontsize=13,color="g")

plt.show();
````

Yukarıdaki grafikte hangi rengin hangi sınıfa ait olduğu belli olmuyor. Bunun için bir "legend" eklememiz gerekiyor. Öncelikle veri setinde kaç sınıf olduğuna bakalım:

````python
iris.target_names
````

Gördüğünüz gibi 3 sınıf var. İlk sınıfı kırmızı, ikinci sınıfı mavi, üçüncü sınıfı yeşil çizdirelim. Bunun için bir for döngüsü kullanırsak iyi olacak. Böylelikle "legend"ı renklere göre atayabiliriz.

````python
# Renk isimlerini belirliyoruz
colors = ["r","b","g"]
# figure oluşturalım
plt.figure(figsize=fig_size)

# x ve y eksenine isimleri de yazalım
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

# for döndürelim
# burada enumerate kalıbı i değişkenini almamıza yarıyor
for i,name in enumerate(iris.target_names):
    if (i <3):
        print(name)
        plt.scatter(iris.data[iris.target==i,0],iris.data[iris.target==i,1],s=iris.data[iris.target==i,2]*50,c=colors[i])
        
plt.legend(iris.target_names)

plt.show()
````

Grafik fena değil ama "legend kısmındaki boyutların homojen olmasını istersek ne yapacağız" dediğinizi duyar gibiyim. Biraz daha kod üstünde oynayalım. Ve son bonus olarak şekilleri de çok ilginç şeyler yapalım :) Bunun için standart denklem formatını kullanacağız.

````python
# Renk isimlerini belirliyoruz
colors = ["r","b","g"]
# şekiller standart denklem formatında
shapes = [r'$\alpha$',r'$\beta$',r'$\gamma$']
# figure oluşturalım
plt.figure(figsize=fig_size)


# x ve y eksenine isimleri de yazalım
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

# for döndürelim
# burada enumerate kalıbı i değişkenini almamıza yarıyor
for i,name in enumerate(iris.target_names):
    plt.scatter(iris.data[iris.target==i,0],iris.data[iris.target==i,1],s=iris.data[iris.target==i,2]*70,
                c=colors[i],
                marker=shapes[i],
               alpha=0.3)
lgnd = plt.legend(iris.target_names)

#Her üç eleman için de boyutu sabitleyelim
lgnd.legendHandles[0]._sizes = [70]
lgnd.legendHandles[1]._sizes = [70]
lgnd.legendHandles[2]._sizes = [70]
````

Sanırım dağılım grafiği için temel fikri anladınız. Şimdi başka bir grafik tipine geçebiliriz.

</div>

<div class="alert alert-success">
Sütun grafikleri üst üste veya yan yana koyarak da görselleştirme yapılabilir.

````python
plt.figure(figsize=fig_size)
y_1 = [10,5,3,12,20]
y_2 = [5,8,9,10,11]
x_labels = ['Ocak','Şubat','Mart','Nisan','Mayıs']
x_pos = np.arange(len(y))

plt.bar(x_pos-0.2,y_1,color="r",width=0.3)
plt.bar(x_pos+0.2,y_2,width=0.3)


plt.title("Satış Grafiği")

plt.xticks(x_pos,x_labels);

plt.xlabel("Aylar")
plt.ylabel("Satış miktarı (x1000)")

plt.legend(["A ürünü","B ürünü"])
plt.show();
````

Yığın halinde sütun çizdirmek istersek "bottom" parametresini bir önceki seri olarak verebiliriz.

````python
plt.figure(figsize=fig_size)
y_1 = [10,5,3,12,20]
y_2 = [5,8,9,10,11]
x_labels = ['Ocak','Şubat','Mart','Nisan','Mayıs']
x_pos = np.arange(len(y))

plt.bar(x_pos,y_1,color="r")
plt.bar(x_pos,y_2,color="b",bottom=y_1)


plt.title("Satış Grafiği")

plt.xticks(x_pos,x_labels);

plt.xlabel("Aylar")
plt.ylabel("Satış miktarı (x1000)")

plt.legend(["A ürünü","B ürünü"])
plt.show();
````

Bir "boxplot" grafiği değişkenin dağılımına ilişkin önemli bilgiler verir. Dağılımın "median", "25th quartile", "50th quartile" değerlerini grafik üzerinden görebiliriz

````python
y = [172,175,181,179,165,167,192,185,155,175,210,115]
# y = np.random.randn(10000) 

plt.figure(figsize=[12,8])

ret = plt.boxplot(y);

#---------------------------
# Annotations
median = [item.get_ydata()[0] for item in ret['medians']]
plt.annotate("Medyan",[0.9,median[0]],[0.6,median[0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")

boxes = [item.get_ydata() for item in ret['boxes']]
plt.annotate("Q3:25th quartile",[0.9,boxes[0][0]],[0.6,boxes[0][0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")
plt.annotate("Q1:75th quartile",[0.9,boxes[0][2]],[0.6,boxes[0][2]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")

caps = [item.get_ydata() for item in ret['caps']]
plt.annotate("Q3-1.5xIQR",[0.95,caps[0][0]],[0.6,caps[0][0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")
plt.annotate("Q1+1.5xIQR",[0.95,caps[1][0]],[0.6,caps[1][0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")
#---------------------------

plt.title("Boxplot örneği")
plt.xticks([],[])
plt.show();
````

## Çizdirme için nesne konseptlerini kullanma

*matplotlib* kütüphanesi OOP (Object Oriented Programming: Nesneye Yönelik Programlama) konseptleri göz önünde bulundurularak geliştirilmiştir. Yani grafik üzerinde göreceğiniz her bir eleman aslında bir **nesnedir.** 

Oluşturduğunuz figür, eksen, çizdiğiniz "line plot", "scatter plot", eksendeki tikler, yazılar hepsi ama hepsi nesnedir. 

Bu nesnelerin içinde birçok özellik vardır. Örneğin bir "line plot"'un rengi, büyüklüğü, stili hep bu "line plot" nesnesinin özelliğidir. Bu nesneyi uygun bir değişkene aktardığınız takdirde bunları değiştirebilmek oldukça kolaylaşır. 

OOP konseptlerine yabancıysanız şimdiden kafanız karışmış olabilir farkındayım, çok normal, benim hala bile bazı durumlarda kafam karışıyor. 

**Neyse ki kafa karışıklığına iyi gelen bir ilacımız, o da bol bol örnek yapmak :)** O yüzden bu yazımızda birçok örnek üzerinden konuyu anlatmaya çalıştım. Sorunuz olursa bana hiç çekinmeden gönderebilirsiniz.

### *matplotlib* ile "klasik" çizdirme yöntemi

Önce çoğumuzun daha önce karşılaştığı, ilk yazımızda da anlattığımız yöntem üzerinde bir gösterim yapalım. *matplotlib* kullanılırken çoğunlukla aşağıdaki gibi bir yöntem uygulanır, çünkü bu yöntem daha kısa satır sayısı ile daha çok işlem yapabilmenize olanak sağlamaktadır.

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 10)
y = x ** 2
plt.plot(x,y)
plt.title("Genel Yöntem")
plt.show()
````

Bu yöntem hızlı bir sonuç almak istediğinizde oldukça yeterlidir. ``plt.plot`` komutunu çağırdığınız zaman matplotlib arkaplanda sizin görmediğiniz birçok işlem yaparak **grafikteki nesneleri en uygun şekilde otomatik ayarlamaktadır.**

````python
# Her seferinde plt.show() demekten kurtuluyoruz ve çeşitli özelliklere sahip oluyoruz
%matplotlib notebook
plt.plot(x,y)
````

Eğer grafiğinize **"ince ayar" yapmak istiyorsanız**, bu arkada koşan nesneleri kullanabilmeyi bilmeniz lazım.

## Örnekler: Nesne tabanlı çizdirme yaklaşımı

Örnek yaparak daha iyi anlayacağınızı düşünüyorum vakit kaybetmeden başlayalım.

### Örnek 1: axes nesnesini kullanma

**axes** en önemli nesnelerden biridir, çünkü çoğu grafik elemanı bu nesnenin içinde yer alır. O yüzden bu kısma biraz daha dikkat etmenizi ve örneği kendinizin de tekrarlamasını istiyorum.

### Örnek 6: Eksenlere ve legend kısmına *LaTeX* ile denklem yazmak

$LaTeX$ genel tanımıyla bir dokümantasyon hazırlama sistemidir. Burada biz bu sistemi güzel denklemler yazmak için kullanağız. *matplotlib* LaTeX konusunda gerçekten iyidir. `$ denklem buraya $` şeklinde istediğiniz denklemi eklemeniz mümkün. LaTeX yazımı çok ayrıntılı bir konu olduğu için bu kısımda sadece örnekteki kullanımları inceleseniz yeterli. 

$LaTeX$ hakkında ayrıntılı bilgi için:<br>
https://matplotlib.org/users/mathtext.html

>**İpucu:** Denklemleri yazdığınız string'in başına r (raw) eklemeyi unutmayın. Eklemezseniz bazen çalışmayabilir.

x'i çeşitli denklemlerde kullanıp, legend kısmına ekleyelim

````python
fig, ax = plt.subplots(figsize = (10,10))

# x ve denklemler
x = np.linspace(1,10,50)
y0 = x
y1 = x**2-10*np.sqrt(x)
y2 = np.log(x)
y3 = 10*np.sin(x/(np.pi))

# $ işareti kullanarak denklem yazıyoruz
# Dikkat ederseniz her bir denklemin etiketini kendi içinde tanımlıyoruz.
ax.plot(x,y0, marker="x",linestyle="--", label = r"$y_0 = x$")
ax.plot(x,y1, marker="o",linestyle="-.", label = r"$y_1 = x^2-10\sqrt{x}$")
ax.plot(x,y2, marker="v",linestyle=":", label = r"$y_2 = \ln(x)$")
ax.plot(x,y3, marker="s",linestyle="-", label = r"$y_3 = 10sin{\frac{x}{\pi}} $" )

# Biraz "legend" boyutunu büyütelim
ax.legend(loc = 2, prop={'size': 20})

# y eksenini logartimic yapalım
ax.set_yscale("log")

# grid ekleyelim ve rengini farklı yapalım
ax.grid(color="m",alpha=0.3,linestyle=":",linewidth=5)

# x ve y eksenindeki tick değerlerini değiştirelim
ax.set_xticks([2,4,6,8])
ax.set_xticklabels([r'x=2',r'x=4',r'x=6',r'x=8'],fontsize=20);

ax.set_yticks([1,10])
ax.set_yticklabels([r'y=1',r'y=10'],fontsize=20);
````

### Örnek 7: Çift y ekseni kullanımı

Bazı durumlarda x'e bağlı iki değişkeni aynı grafikte göstermek çok zor olabilir. Örneğin birisi üssel olarak artıyor diğeri lineer artıyorsa bunların üssel olanın değeri çok büyük diğeri ise ona kıyasla çok çok küçüktür. Bu aşamda imdadımıza çift eksen kullanımı yetişecektir.

>**İpucu:** Çift eksen çizdirme ``twinx`` fonksiyonu kullanılarak yapılmaktadır. İsterseniz ```twiny`` da mevcut

````python
fig, ax = plt.subplots()

# x ve denklemler
x = np.linspace(1,10,50)
# lineer
y1 = x
# üssel
y2 = np.exp(x)
# ilk ekseni çizdirelim
ax.plot(x,y1,"b")

# ilk y eksenin etiketini yazalım
ax.set_ylabel("$y=x$",color="b",fontsize=15)
# # y eksenindeki fontu da modifye edelim
for label in ax.get_yticklabels():
    label.set_color("blue")
    label.set_fontsize(12)
    
# # eksenin ikizini oluşturalım
ax_exp = ax.twinx()

# # ikinci ekseni çizdirelim
ax_exp.plot(x,y2,"r")
# ikinci y eksenin etiketini yazalım
ax_exp.set_ylabel("$y=e^x$",color="r",fontsize=15)
for label in ax_exp.get_yticklabels():
    label.set_color("red")
    label.set_fontsize(12)
````

### Örnek 8: İki çizgi grafiğin arasını `fill_between` fonksiyonu ile doldurma

y ekseninde iki fonsiyonunuz var ve bunların arasını doldurmak istiyorusunuz öyle mi? Challenge accepted!!!

### Örnek 9: `polar` grafik çizme

Çok ihtiyacınız olmayabilir fakat nasıl yapıldığına mutlaka bir göz atın bence. 

>**İpucu:** Ekseni `polar` olarak tanımladıktan sonra grafik tipini istediğiniz gibi seçebilirsiniz.

````python
# Burada subplot metodunu kullanmıyoruz
fig = plt.figure(figsize = (10,10))
# add_plot yerine add_subplot kullandık. 
# Bunun avantajı add_subplot'taki 
# gibi sınır tanımlamamıza gerek olmamasıdır 
ax = fig.add_subplot(111,projection='polar')

# eksenler r ve theta olacak
N=150
# Rasgele sayılar üretelim 
r = 3*np.random.rand(N)
# theta 
theta = 2 * np.pi * np.random.rand(N)

# bir spiral çizdirelim
theta2 = np.linspace(0,2*np.pi,30)

# scatter ekleyelim
# theta'ya göre renk verelim, biraz da alpha
# Boyutunu da 200*r**2 yapalım
ax.scatter(theta,r,
           c=theta,
           alpha=0.5,
           edgecolor="k",
           s=r**2*200,
           cmap="jet",label="scatter");

# # spiral ekleyelim
ax.plot(theta2,theta2/2,"r",marker="o",label="spiral");

# # Histogram da olur mu dersiniz?
var = np.random.uniform(low=0,high=2*np.pi,size=20)
ax.hist(var,color="r",alpha=0.2,edgecolor="k",linewidth=2,label="hist");

ax.legend()
````

### Örnek 11: Kontür bulma, `pcolor` gösterimi ve yatay `colorbar` çizdirme

Daha öne haritalarda eşdeğer çizgiler görmüşsünüzdür. Kontür bulma bu çizgilerin iki boyutlu düzlemde gösterilmesidir. Kontür bulmayı genelikle üç boyutlu veriler üzerinde yaparız. Bunun için ilk önce 2 boyutlu bir ızgara (grid) yaratalım ve bu ızgaranın üzerine bir fonksiyon oturtalım.

````python
%matplotlib inline

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
# ızgarayı oluşturalım.
X, Y = np.meshgrid(x, y)

# iki değerin karesinin toplamını alalım
Z = X**2 + Y**2

#yeni figure oluşturuyoruz
# fig,ax = plt.subplots()
# # contour çizdirelim
# CS = ax.contour(X, Y, Z,levels = np.arange(0,14,1),cmap="jet")
# # Bu çizdiriğimiz kontor değerlerine  etiket atayalım
# plt.clabel(CS, inline=1, fontsize=10)

# # aynı örneği bir de renkli olarak gösterelim
fig1,ax1 = plt.subplots()
out = ax1.pcolor(X,Y,Z)
# bir colorbar ekleyelim
# colorbar'ın kendi ekseni vardır
col_bar = fig1.add_axes([0.2, 0, 0.5, 0.05])
fig1.colorbar(out,cax=col_bar, orientation='horizontal')
````

### Örnek 12: Üç boyutlu çizim

*matplotlib* kütüphanesinin üç boyutlu çizim için en iyi ortam olduğunu söylemek biraz güç. 

Üç boyutlu çizim yapabilmek için öncelikle aşağıdaki kütüphaneyi ``import`` etmemiz lazım.

````python
from mpl_toolkits.mplot3d.axes3d import Axes3D
````

Şimdi üç boyutlu bir yüzey çizimi yapalım, polar grafik çiziminde olduğu gibi bunda da `projection` parametresini ayarlamamız gerekiyor.

````python
%matplotlib notebook

fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(111,projection='3d')

# Veri hazırlayalım
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
# ızgarayı oluşturalım.
X, Y = np.meshgrid(x, y)
# iki değerin karesinin toplamını alalım
Z = X**2 + Y**2

# cool bir renk seçimi olsun :)
surf = ax.plot_surface(X,Y,Z,cmap="jet");
#colorbar ekleyelim ve biraz küçültelim shrink = 0.6 olsun
fig.colorbar(surf,shrink=0.6)
````

### Örnek 13: Üç boyutlu çizim üzerinde `contour` gösterimi ve bakış açısını değiştirme

````python
fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(111,projection='3d')

# Veri hazırlayalım
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
# ızgarayı oluşturalım.
X, Y = np.meshgrid(x, y)
# iki değerin karesinin toplamını alalım
Z = X**2 + Y**2

# cool bir renk seçimi olsun :)
surf = ax.plot_surface(X,Y,Z,cmap="cool");
#colorbar ekleyelim ve biraz küçültelim shrink = 0.6 olsun
fig.colorbar(surf,shrink=0.6)

ax.contour(X, Y, Z, zdir='z', offset=-0, cmap="gist_rainbow")
ax.contour(X, Y, Z, zdir='x', offset=-3, cmap="Pastel2_r")
ax.contour(X, Y, Z, zdir='y', offset=-3, cmap="coolwarm")

# Biraz görünümü de değiştirelim
ax.view_init(90,90)
````
