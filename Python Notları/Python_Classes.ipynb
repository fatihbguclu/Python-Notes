### Inheritance (Kalıtım / Miras Alma)

Kalıtım, bir sınıfın diğerinin niteliklerini ve yöntemlerini aldığı süreçtir. Yeni oluşturulan sınıflara alt sınıflar, alt sınıfların türetildiği sınıflara da ana sınıflar denir.

Alt sınıflar ebeveynin tüm nitelik ve davranışlarını miras alır, ancak takip edilecek farklı davranışları da belirtebilir.

```python
# Temel hayvan sinifi
class Hayvan:
    
    canli = True
    hareket_edebilir = True
    cok_hucreli = True
    hucre_duvari_var = False
    
    def beslenme(self):
        pass
    
    def buyume(self):
        print("Hayvanlar zamanla büyür...")
        pass
    
    def hareket(self):
        pass
    
    def ureme(self):
        pass
    
    def disari_tepki(self):
        pass
```

```python
class Kedi(Hayvan):
    
    # Tum kediler icin 
    tur = "Kedi"
    
    def __init__(self,isim,renk):
        self.isim = isim
        self.renk = renk
        
    def miyavla(self):
        print("Miyavvv!")
```

```python
a = Kedi(isim = "Garfield",renk="turuncu")
a.buyume()
print(a.isim + "\n" + a.renk)
a.miyavla()
```

### Operator overloading

```python
class Nokta:
    
    def __init__(self,x=0,y=0):
        self.x = x
        self.y = y
        
        
a = Nokta(1,1)
b = Nokta(2,2)

a + b
```

```python
class Nokta:
    
    def __init__(self,x=0,y=0):
        self.x = x
        self.y = y
        
    def __add__(self,other):
        x = self.x + other.x
        y = self.y + other.y
        
        return Nokta(x,y)
    
    def __str__(self):
        return "Noktanın Pozisyonu ({},{})".format(self.x,self.y)
    
a = Nokta(2,4)
b = Nokta(5,-1)
c = a + b
print(c)
```

>Hangi operatörleri "overload" yapabiliriz? 

Addition p1 + p2 p1.add(p2)

Subtraction p1 - p2 p1.sub(p2)

Multiplication p1 * p2 p1.mul(p2)

Power p1 ** p2 p1.pow(p2)

Division p1 / p2 p1.truediv(p2)

Floor Division p1 // p2 p1.floordiv(p2)

Remainder (modulo) p1 % p2 p1.mod(p2)

Bitwise Left Shift p1 << p2 p1.lshift(p2)

Bitwise Right Shift p1 >> p2 p1.rshift(p2)

Bitwise AND p1 & p2 p1.and(p2)

Bitwise OR p1 | p2 p1.or(p2)

Bitwise XOR p1 ^ p2 p1.xor(p2)

Bitwise NOT ~p1 p1.invert()

Less than	p1 < p2	p1.__lt__(p2)

Less than or equal to	p1 <= p2	p1.__le__(p2)

Equal to p1 == p2	p1.__eq__(p2)

Not equal to	p1 != p2	p1.__ne__(p2)

Greater than	p1 > p2	p1.__gt__(p2)

Greater than or equal to	p1 >= p2	p1.__ge__(p2)

### super class kullanımı

```python
# Temel hayvan sinifi
class Hayvan:
    
    canli = True
    hareket_edebilir = True
    cok_hucreli = True
    hucre_duvari_var = False
    
    def __init__(self,yas,kilo,boy): #Üst Class Constr
        self.yas = yas
        self.kilo = kilo
        self.boy = boy
    
    def beslenme(self):
        pass
    
    def buyume(self):
        print("Hayvanlar zamanla büyür...")
        pass
    
    def hareket(self):
        pass
    
    def ureme(self):
        pass
    
    def disari_tepki(self):
        pass
```

```python
class Kus(Hayvan):
    
    tur = "Kus"
    
    def __init__(self,yas,kilo,boy,renk,isim): #alt Class Constr
        super().__init__(yas,kilo,boy)
        self.isim = isim
        self.renk = renk
    
    def ses_cikar(self):
        print("Cik cik cik...")
```

```python
k = Kus(10,10,10,"sarı","tweety")

k.ses_cikar()

k.kilo
```

### Çoklu "inheritance"

Birden fazla sınıfın özelliklerini ve metodlarını devralmak demektir

```python
class Muzisyen:
    def __init__(self,enstruman):
        self.enstruman = enstruman
    
    def enstruman_cal(self):
        print(self.enstruman + " sesi geliyor")
        
class Muhendis:
    def __init__(self,brans):
        self.brans = brans
    
    def hesaplama_yap(self):
        print("hesaplama yapılıyor")
        
```

```python
class Muzisyen_Muhendis(Muzisyen,Muhendis):
    def __init__(self,brans,enstruman):
        Muzisyen.__init__(self,enstruman=enstruman)
        Muhendis.__init__(self,brans = brans)
```

```python
a = Muzisyen_Muhendis(brans="EE",enstruman="piyano")
a.enstruman_cal()
a.hesaplama_yap()
```

### Polymorphism (Çok biçimlilik)

Çoklu veri tipleri için tek bir arayüz kullanmak demektir.

Birkaç sınıf veya alt sınıf aynı yöntem adına sahipse, ancak bu aynı yöntemler farklı işlevlere sahipse, sınıflar polimorfiktir, çünkü farklı türdeki tek bir arabirim kullanarak farklı işlemler yapılabilir. Bir fonksiyon, hangi sınıfın çağrıldığını bilmeden aynı işlemi yapabilir.

```python
class Balik:
    pass

class Sazan(Balik):
    def yuz(self):
        print("Sazan yüzüyor")
        
class Levrek(Balik):
    def yuz(self):
        print("Levrek Yüzüyor")
        
```

```python
def balik_yuz(obj):
    obj.yuz()
```

```python
levrek = Levrek()
sazan = Sazan()
balik_yuz(levrek)
balik_yuz(sazan)
```

### Abstraction (Soyutlama)

Ust seviye siniflar bir sablon niteliğinde olmali

Ust seviyede yazilan siniflarin kendinden turetilecek siniflarda bazi metotlari "implement" etmesini zorunlu olmalıdır

Bunun icin "abstract base classes" abc kutuphanesini import etmek gerekecek

```python
from abc import ABCMeta, abstractmethod

class Hayvan(metaclass=ABCMeta):
    
    @abstractmethod
    def yuru(self):
        pass
    
    @abstractmethod
    def ses_cikar(self):
        pass
    
```

```python
class Kopek(Hayvan):
    
    def __init__(self,isim):
        print("Köpek Oluşturuldu")
        self.isim = isim
        
    def yuru(self):
        print(self.isim + " Yürüyor.")
    
    def ses_cikar(self):
        print("hav hav hav")
        
        
        
a = Kopek("paşa")
a.yuru()
a.ses_cikar()
```
