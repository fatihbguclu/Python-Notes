* df.groupy()
    * Çok faydalı bir fonksiyondur. Bir sütunu gruplayıp üzerinde bir operasyon yaptıktan sonra özet tablo oluşturabilirsiniz.

```python
ex = {"isim":["Ali","Veli","Ayşe","Ahmet","Burcu","Cansu","Ramiz"],"cinsiyet":["E","E","K","E","K","K","E"],"boy":[180,160,170,150,170,200,190],"kilo":[80,50,55,60,50,110,90],"yaş":[15,15,40,72,65,20,27]}
df_example = pd.DataFrame(ex)
df_example
```

```python
# Cinsiyet üzerinden gruplandırıp ortalamasını bulalım
df_example.groupby("cinsiyet").mean()
```

```python
# Hangi sütunların gruplandırılacağını da seçebilirsiniz.
# Burada "max" kullanalım değişiklik olsun
df_example.groupby("cinsiyet")[["boy","yaş"]].max()
```

* Pivot tablosu oluşturalım

```python
df_example.pivot_table(index = ["cinsiyet","yaş"],aggfunc = "max")
```

<div class="alert alert-warning">
<b> Dataframe üzerinde "custom" fonksiyon kullanımı  </b> 
</div>

```python
# dikdortgen
dikdortgenler = [
    { 'en': 40, 'boy': 10 },
    { 'en': 20, 'boy': 25 },
    { 'en': 10, 'boy': 15 }
]

dikdortgenler_df = pd.DataFrame(dikdortgenler)
dikdortgenler_df
```

```python
def alan_hesapla(satir):
    return satir["en"]*satir["boy"]

dikdortgenler_df["alan_metot1"] = dikdortgenler_df.apply(alan_hesapla,axis=1)
dikdortgenler_df
```

```python
dikdortgenler_df["alan_metot2"] = dikdortgenler_df.apply(lambda satir:satir["en"]*satir["boy"],axis=1)
dikdortgenler_df
```

```python
age_pd_series = pd.Series({"A":33,"B":24,"C":45,"E":18},name="yaş")
weight_pd_series = pd.Series({"A":55,"B":65,"C":75,"D":80},name="kilo") # in kg :)

custom_df = pd.concat([age_pd_series,weight_pd_series],axis=1,sort=True)
custom_df
```

```python
custom_df.apply(lambda sutun:sutun.max()-sutun.min(),axis=0)
```

**`apply` sütunlar veya satırlar üzerinde işlem yapar. `applymap` elemanlar üzerinde işlem yapar.**

```python
# Her elemanın karesini alalım
custom_df.applymap(lambda x:x**2)
```
